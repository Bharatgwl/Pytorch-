# Full Pytorch training pipeline


```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import torch.nn as nn
from torch.utils.data import Dataset ,DataLoader
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 32','id'],inplace=True)
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42   )
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

```python
X_test_tensor.shape
```

```python
class customdataset(Dataset):
  def __init__(self, features, output_labels):
    self.features = features
    self.output_labels = output_labels
  def __len__(self):
    return self.features.shape[0]
  def __getitem__(self, index):
    return self.features[index], self.output_labels[index]
```

```python
train_dataset = customdataset(X_train_tensor,y_train_tensor)
test_dataset = customdataset(X_test_tensor,y_test_tensor)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32,shuffle=True)
```

```python
class NNN(nn.Module):
  def __init__(self, input_features):
    super().__init__()
    self.linear = nn.Linear(input_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    out = self.linear(x)
    out = self.sigmoid(out)
    return out
```

```python
epoch = 30
lr = 0.1
loss_function = nn.BCELoss()
model = NNN(X_train_tensor.shape[1])
optim = torch.optim.SGD(model.parameters(), lr = lr)
```

```python
def train():
  for e in range(epoch):
    for batch_features , batch_labels in train_loader:
      y_pred  = model(batch_features) # forward pass
      loss = loss_function(y_pred, batch_labels.view(-1,1).float()) # loss calc
      optim.zero_grad() # clear old grad
      loss.backward()# backward pass
      optim.step() # update params
    print(f"epoch:{e+1} , loss:{loss.item()}")

```

```python
train()
```
