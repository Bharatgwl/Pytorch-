```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import torch.nn as nn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 32','id'],inplace=True)
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.FloatTensor(X_train)
X_test_tensor = torch.FloatTensor(X_test)
y_train_tensor = torch.LongTensor(y_train)
y_test_tensor = torch.LongTensor(y_test)
```

```python
class MysimpleNN(nn.Module):
  def __init__(self,input_features):
    super().__init__()
    self.linear1 = nn.Linear(input_features,1)
    self.sigmoid = nn.Sigmoid()
  def forward(self,x):
    out = self.linear1(x)
    out = self.sigmoid(out)
    return out
  def loss(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    loss = -(y * torch.log(y_pred) +
             (1 - y) * torch.log(1 - y_pred)).mean()

    return loss
```

```python
lr= 0.1
epochs = 25

```

```python
model = MysimpleNN(X_train_tensor.shape[1])
```

```python
for epoch in range(epochs):

  ypred = model.forward(X_train_tensor)
  loss = model.loss(ypred,y_train_tensor)

  loss.backward()

  with torch.no_grad():
    model.linear1.weight-=lr*model.linear1.weight.grad
    model.linear1.bias-=lr*model.linear1.bias.grad

  model.linear1.weight.grad.zero_()
  model.linear1.bias.grad.zero_()

  print(f"epochs : {epoch+1} loss : {loss}")
```

```python
ypred = model.forward(X_train_tensor)
```

```python
model.loss(ypred,y_train_tensor)
```

```python
with torch.no_grad():
  ypred = model.forward(X_test_tensor)
  ypred = (ypred>0.9).float()
  accuracy = (ypred == y_test_tensor).float().mean()
  print(f"accuracy : {accuracy}")
```

```python
class MysimpleNN(nn.Module):
  def __init__(self,input_features):
    super().__init__()
    self.linear1 = nn.Linear(input_features,1)
    self.sigmoid = nn.Sigmoid()
  def forward(self,x):
    out = self.linear1(x)
    out = self.sigmoid(out)
    return out
```

```python
lr= 0.1
epochs = 25

loss_function = nn.BCELoss()
```

```python
type(loss_function)
```

```python
model = MysimpleNN(X_train_tensor.shape[1])
optimiser = torch.optim.SGD(model.parameters(),lr=lr)
```

```python
for epoch in range(epochs):
  # forward pass
  ypred = model(X_train_tensor)
  # loss calc
  loss = loss_function(ypred, y_train_tensor.reshape(-1,1).float())
  # clear old grads
  optimiser.zero_grad()
  # backprop
  loss.backward()
  # param update
  optimiser.step()

  print(f"epochs : {epoch+1} loss : {loss}")
```

```python
with torch.no_grad():
  ypred = model.forward(X_test_tensor)
  ypred = (ypred>0.9).float()
  accuracy = (ypred == y_test_tensor).float().mean()
  print(f"accuracy : {accuracy}")
```
