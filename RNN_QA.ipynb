```python
# tokenize
def tokenize(text):
    text = text.lower()
    text = text.replace("?","")
    text = text.replace("'","")
    return text.split()
```

```python
tokenize(df["question"][0])
```

```python
# vocab 

vocab = {'<UNK>':0}
```

```python
def build_vocab(texts):
    tokenized_question = tokenize(texts['question'])
    tokenized_ans = tokenize(texts['answer'])
    
    merged = tokenized_question + tokenized_ans
    print(merged)
    
    for word in merged:
        if word not in vocab:
            vocab[word] = len(vocab)
```

```python
df.apply(build_vocab, axis=1)
```

```python
vocab
```

```python
# convert to index
def text_to_index(text,vocab):
    indexe_text = []
    
    for word in tokenize(text):
        
        if word in vocab:
            indexe_text.append(vocab[word])
        else:
            indexe_text.append(vocab['<UNK>'])
            
    return indexe_text
    
```

```python
text = "What is the capital of cam?"
text_to_index(text,vocab)
```

```python
import torch 
from torch.utils.data import Dataset, DataLoader
```

```python
class QADataset(Dataset):
    def __init__(self, df, vocab):
        self.df = df
        self.vocab = vocab
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        question = self.df.iloc[idx]['question']
        answer = self.df.iloc[idx]['answer']
        
        question_indexed = text_to_index(question,self.vocab)
        answer_indexed = text_to_index(answer,self.vocab)
        
        return torch.tensor(question_indexed),torch.tensor(answer_indexed)
        
```

```python
dataset = QADataset(df,vocab)
```

```python
dataset[0]
```

```python
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
for question, answer in dataloader:
    print(question, answer)
    
    
```

```python
class MysimpleRNN(torch.nn.Module):
    def __init__(self, vocab_size):
        super(MysimpleRNN, self).__init__()
        self.vocab_size = vocab_size
        self.embedding = torch.nn.Embedding(vocab_size, embedding_dim=50)
        self.rnn = torch.nn.RNN(input_size = 50 , hidden_size = 50, batch_first = True) 
        self.fc = torch.nn.Linear(50, vocab_size)
    def forward(self, x):
        x = self.embedding(x)
        hidden, final = self.rnn(x)
        output = self.fc(final.squeeze(0))
        return output
```

```python
learning_rate = 0.001
epochs = 30 
criterion = torch.nn.CrossEntropyLoss()
model = MysimpleRNN(len(vocab)).to(torch.device("cuda"))
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for epoch in range(epochs):
    total_loss = 0
    for question ,answer in dataloader:
        question = question.to(torch.device("cuda"))
        answer = answer.to(torch.device("cuda"))
        
        output = model(question)
        loss = criterion(output, answer.squeeze(1))
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
def pred(model, question, vocab):
    model.eval()
    with torch.no_grad():
        question_indexed = text_to_index(question, vocab)
        question_tensor = torch.tensor(question_indexed).unsqueeze(0).to(torch.device("cuda")) # unsqueeze to add batch dimension
        output = model(question_tensor)
        probs = torch.nn.functional.softmax(output, dim=1)
        
        values, index = torch.max(probs, dim=1)
        
        if values.item() < 0.5:  # Threshold for confidence
            return "I don't know the answer to that question."
        else:
            for word, idx in vocab.items():
                if idx == index.item():
                    return word
```

```python
pred(model, "What is the largest planet in our solar system", vocab)
```
