```python
class Atm:
    def __init__(self):
        self.balance = 0
        self.pin = ""
        self.menu()

    def menu(self):
        user_input = input(
            """
        Welcome to the ATM Machine!
        Please choose an option:
        1. Deposit
        2. Withdraw
        3. Check Balance
        4. Create pin
        5. Exit
        """
        )

        if user_input == "1":
            self.deposit()
        elif user_input == "2":
            self.withdraw()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.create_pin()
        elif user_input == "5":
            print("Thank you for using the ATM Machine. Goodbye!")
            exit()
        else:
            print("Invalid option, please try again.")
            self.menu()

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Invalid amount, please enter a positive number.")
        self.menu()

    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Invalid amount, please enter a positive number.")
        self.menu()

    def create_pin(self):
        new_pin = input("Enter a new 4-digit PIN: ")
        if len(new_pin) == 4 and new_pin.isdigit():
            self.pin = new_pin
            print("PIN created successfully.")
        else:
            print("Invalid PIN. Please enter a 4-digit number.")
        self.menu()

    def check_balance(self):
        if self.pin:
            entered_pin = input("Enter your PIN to check balance: ")
            if entered_pin == self.pin:
                print(f"Your current balance is: ${self.balance:.2f}")
            else:
                print("Incorrect PIN.")
        else:
            print("No PIN set. Please create a PIN first.")
        self.menu()
```

## Fractions class 3/4

```python
class Fractions:
    def __init__(self, x, y):
        self.num = x
        self.den = y

    def __str__(self):
        return f"{self.num}/{self.den}"

    def __add__(self, other):
        new_num = self.num * other.den + self.den * other.num
        new_den = self.den * other.den

        return f"{new_num}/{new_den}"

    def __sub__(self, other):
        new_num = self.num * other.den - self.den * other.num
        new_den = self.den * other.den

        return f"{new_num}/{new_den}"
    def __mul__(self, other):
        new_num = self.num * other.num 
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    def __truediv__(self, other):
        new_num = self.num * other.den 
        new_den = self.den * other.num
        return f"{new_num}/{new_den}"   
    def to_dec(self):
        return self.num/self.den
```

## Coordinate geometry class where we can create point and apply distance based operations like euclidean distance 

```python
class Sample:
    def __init__(self, x):
        self.x = x

    def __str__(self):
        return f"Sample object with value: {self.x}"
    def sq(self, x):
        return f"Value is {x**2}"
    def multiply(self, other):
        return self.x * self.x
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"<{self.x},{self.y}>"

    def Euclidean_distance(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5

    def distance_from_origin(self):
        return self.Euclidean_distance(Point(0, 0))


class Line:

    def __init__(self, p1, p2, p3):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3

    def __str__(self):
        return f"({self.p1})x + ({self.p2})y + {self.p3} = 0"
        # if(self.p1<0):
        #     return f"-{self.p1}x + {self.p2}y + {self.p3} = 0"
        # elif(self.p2<0):
        #     return f"{self.p1}x - {self.p2}y + {self.p3} = 0"
        # elif(self.p3<0):
        #     return f"{self.p1}x + {self.p2}y - {self.p3} = 0"
        # elif(self.p1<0 and self.p2<0):
        #     return f"-{self.p1}x - {self.p2}y + {self.p3} = 0"
        # else:
        #     return f"{self.p1}x + {self.p2}y + {self.p3} = 0"

    def point_on_line(self, point):
        return self.p1 * point.x + self.p2 * point.y + self.p3 == 0

    def shortest_distance_from_point(self, point):
        return abs(self.p1 * point.x + self.p2 * point.y + self.p3) / (
            (self.p1**2 + self.p2**2) ** 0.5
        )

    def line_intersection(self, other):
        determinant = self.p1 * other.p2 - other.p1 * self.p2
        if determinant == 0:
            return "Lines are parallel"
        else:
            x = (other.p2 * (-self.p3) - self.p2 * (-other.p3)) / determinant
            y = (self.p1 * (-other.p3) - other.p1 * (-self.p3)) / determinant
            return Point(x, y)
```

```python
class Atm:
    def __init__(self):
        self.__balance = 0      # private variable
        self.pin = ""
        self.menu()

    # -------- Getter --------
    def get_balance(self):
        return self.__balance

    # -------- Setter --------
    def set_balance(self, amount):
        if amount >= 0 and isinstance(amount, (int, float)):
            self.__balance = amount
        else:
            print("Balance cannot be negative!")

    def menu(self):
        user_input = input(
            """
        Welcome to the ATM Machine!
        Please choose an option:
        1. Deposit
        2. Withdraw
        3. Check Balance
        4. Create pin
        5. Exit
        """
        )

        if user_input == "1":
            self.deposit()
        elif user_input == "2":
            self.withdraw()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.create_pin()
        elif user_input == "5":
            print("Thank you for using the ATM Machine. Goodbye!")
            exit()
        else:
            print("Invalid option, please try again.")
            self.menu()

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.set_balance(self.get_balance() + amount)
            print(f"Deposited ${amount:.2f}. New balance: ${self.get_balance():.2f}")
        else:
            print("Invalid amount, please enter a positive number.")
        self.menu()

    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if amount > 0:
            if amount <= self.get_balance():
                self.set_balance(self.get_balance() - amount)
                print(f"Withdrew ${amount:.2f}. New balance: ${self.get_balance():.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Invalid amount, please enter a positive number.")
        self.menu()

    def create_pin(self):
        new_pin = input("Enter a new 4-digit PIN: ")
        if len(new_pin) == 4 and new_pin.isdigit():
            self.pin = new_pin
            print("PIN created successfully.")
        else:
            print("Invalid PIN. Please enter a 4-digit number.")
        self.menu()

    def check_balance(self):
        if self.pin:
            entered_pin = input("Enter your PIN to check balance: ")
            if entered_pin == self.pin:
                print(f"Your current balance is: ${self.get_balance():.2f}")
            else:
                print("Incorrect PIN.")
        else:
            print("No PIN set. Please create a PIN first.")
        self.menu()
```

### Super keyword 

### It is used to call the overridden function by child class 

```python
class Parent:
    def __init__(self):
        self.__num = 100
    def show(self):
        print("Parent:",self.__num)
class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__var = 200
    def show(self):
        super().show()
        print("Child:",self.__var)
        
        
```

```python
obj = Child()
obj.show()
```

```python
# Hierarchical Inheritance
class Phone:
    def __init__(self,price,brand,camera):
        print("This is a phone")
        self.__price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying the phone...")
        
class Smartphone(Phone):
    pass
class FeaturePhone(Phone):
    pass
Smartphone(1000,"Apple","12MP").buy()
FeaturePhone(500,"Nokia","2MP").buy()
```

```python
# Multiple Inheritance 
class Phone:
    def __init__(self,price,brand,camera):
        print("This is a phone")
        self.__price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying the phone...")
class Product:
    def review(self):
        print("This is a product")
class Smartphone(Phone,Product):
        pass    

s = Smartphone(1000,"Apple","12MP")
s.buy()
s.review()
```

```python
# the diamond problem
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
class C(A):
    def show(self):
        print("C")
class D(B,C):
    pass
```
